```python
# LabelEncode Categorical Features

from sklearn.preprocessing import LabelEncoder

data = data.apply(LabelEncoder().fit_transform)
```

```python
data.head()
```

```python
data.shape, target.shape
```

```python
# Get target into 1D Array

target = pd.DataFrame(target)

target = target.values.ravel()
```

```python
# Feature Selection

from sklearn.feature_selection import  SelectKBest, f_regression, f_classif
selector = SelectKBest(score_func=f_classif, k=3)

target = target
not_target = data

predictors = not_target

selector.fit(predictors, target)

# selector.fit(Reddit[predictors], Reddit[target].values.ravel())

best_features = selector.get_support(indices=True)

features = list(predictors.columns[selector.get_support(indices = True)])
features
```

```python
# Train-Train-Split on Data Set

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

target = target
not_target = ['education-num', 'relationship', 'capital-gain']

X = data[not_target]
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Standard Scaler

ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
# XGBoost Classifier

gs_params = {
    'max_depth':[1, 2, 3, 4, 5],
    'n_estimators':[1, 5, 10],
    'learning_rate':np.logspace(-5,0,100),
    'silent' : [False],
    'booster' : ['gbtree', 'gblinear', 'dart'] 
}

xgb_gs = GridSearchCV(XGBClassifier(), gs_params, verbose=1)

xgb_gs = xgb_gs.fit(X_train, y_train)

best_xgb_gs = xgb_gs.best_estimator_

xgb_gs_train = best_xgb_gs.score(X_train, y_train)
xgb_gs_test = best_xgb_gs.score(X_test, y_test)

# X_matrix = X.as_matrix()
# y_matrix = y.as_matrix()

# xgb = XGBClassifier()
# xgb.fit(X_matrix, y_matrix)
# perm_xgb = PermutationImportance(xgb).fit(X_matrix, y_matrix)
# eli5.show_weights(perm_xgb, feature_names=list(X))
```
